```python
import os
import fitz
import cv2
import numpy as np
import io
import json
from PIL import Image
from google.colab.patches import cv2_imshow

# 1. フォルダ分けを行ったベースのパスを指定
base_path = '/content/drive/MyDrive/ソリューション開発1&2/フォーマット分類済み'

def extract_boxes_from_image(img, min_area=3000):
    """画像から四角形の座標を抽出する内部関数"""
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15, 5)
    kernel = np.ones((3,3), np.uint8)
    dilated = cv2.dilate(thresh, kernel, iterations=1)
    contours, _ = cv2.findContours(dilated, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    boxes = []
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if min_area < area < (img.shape[0] * img.shape[1] * 0.9):
            epsilon = 0.02 * cv2.arcLength(cnt, True)
            approx = cv2.approxPolyDP(cnt, epsilon, True)
            if len(approx) == 4:
                x, y, w, h = cv2.boundingRect(cnt)
                boxes.append({"x": x, "y": y, "w": w, "h": h})

    # y座標、続いてx座標でソート（左上から右下へ並べるため）
    boxes.sort(key=lambda b: (b['y'], b['x']))
    return boxes

def register_all_templates(base_dir):
    templates_config = {}

    # ベースパス内のフォルダ（フォーマットA, フォーマットB...）を取得
    folders = [f for f in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, f))]

    for folder_name in sorted(folders):
        folder_path = os.path.join(base_dir, folder_name)
        files = [f for f in os.listdir(folder_path) if f.lower().endswith('.pdf') or f.lower().endswith(('.png', '.jpg', '.jpeg'))]

        if not files:
            continue

        # 各フォルダの最初のファイルを「代表テンプレート」として選択
        target_file = files[0]
        target_path = os.path.join(folder_path, target_file)

        print(f"\n{'='*50}")
        print(f"テンプレート登録処理中: {folder_name} (使用ファイル: {target_file})")
        print(f"{'='*50}")

        try:
            # PDFの場合は画像化
            if target_file.lower().endswith('.pdf'):
                doc = fitz.open(target_path)
                page = doc.load_page(0)
                pix = page.get_pixmap(matrix=fitz.Matrix(300/72, 300/72))
                img_bytes = pix.tobytes("png")
                image = Image.open(io.BytesIO(img_bytes)).convert('RGB')
                img = np.array(image)[:, :, ::-1].copy()
                doc.close()
            else:
                img = cv2.imread(target_path)

            # 枠線の検出
            boxes = extract_boxes_from_image(img, min_area=3000)

            # 描画して確認用に出力
            result_img = img.copy()
            for i, b in enumerate(boxes):
                x, y, w, h = b['x'], b['y'], b['w'], b['h']
                cv2.rectangle(result_img, (x, y), (x+w, y+h), (0, 0, 255), 2)
                # 枠の番号を振る
                cv2.putText(result_img, f"No.{i}", (x+5, y+25), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)

                # ★ ここを追加: コンソールに各No.の座標を出力
                print(f"  No.{i} 座標: [X:{x}, Y:{y}, W:{w}, H:{h}]")

            h_img, w_img = result_img.shape[:2]
            cv2_imshow(cv2.resize(result_img, (int(w_img/3), int(h_img/3))))

            # 辞書に保存
            templates_config[folder_name] = {
                "source_file": target_file,
                "regions": boxes
            }
            print(f"-> 計 {len(boxes)} 個の枠を検出・登録しました。")

        except Exception as e:
            print(f"エラーが発生しました: {e}")

    # 全てのテンプレート情報をJSONファイルとして保存
    output_json_path = os.path.join(base_dir, 'templates_config.json')
    with open(output_json_path, 'w', encoding='utf-8') as f:
        json.dump(templates_config, f, ensure_ascii=False, indent=4)

    print(f"\n✅ 全フォーマットのテンプレート設定を保存しました！: {output_json_path}")
    return templates_config

# 実行
all_configs = register_all_templates(base_path)
```

```python
import cv2
import numpy as np
import fitz  # PyMuPDF
import matplotlib.pyplot as plt
from paddleocr import PaddleOCR
from google.colab.patches import cv2_imshow

# --- 1. PDFを画像化する関数 (dpi=300) ---
def pdf_to_cv2_image(pdf_path, page_num=0, dpi=300):
    doc = fitz.open(pdf_path)
    page = doc.load_page(page_num)
    zoom = dpi / 72.0
    mat = fitz.Matrix(zoom, zoom)
    pix = page.get_pixmap(matrix=mat)
    img_array = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.h, pix.w, pix.n)
    if pix.n == 4:
        return cv2.cvtColor(img_array, cv2.COLOR_RGBA2BGR)
    return cv2.cvtColor(img_array, cv2.COLOR_RGB2BGR)

# --- 2. 枠線を検出する関数 (ご提示のコードのロジック) ---
def get_all_boxes(img, min_area=3000):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15, 5)
    kernel = np.ones((3,3), np.uint8)
    dilated = cv2.dilate(thresh, kernel, iterations=1)
    contours, _ = cv2.findContours(dilated, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    boxes = []
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if min_area < area < (img.shape[0] * img.shape[1] * 0.9):
            epsilon = 0.02 * cv2.arcLength(cnt, True)
            approx = cv2.approxPolyDP(cnt, epsilon, True)
            if len(approx) == 4:
                x, y, w, h = cv2.boundingRect(cnt)
                boxes.append({"x": x, "y": y, "w": w, "h": h})

    # Y, Xの順でソートしてNo.を確定させる
    boxes.sort(key=lambda b: (b['y'], b['x']))
    return boxes

# ==========================================
# メイン処理：複数枠の切り取りとOCR
# ==========================================

# 設定
pdf_path = "/content/drive/MyDrive/ソリューション開発1&2/フォーマット分類済み/フォーマットM/MPATFS87JNT000219.pdf"
target_numbers = [40, 41, 42, 43, 44, 45, 46]  # ← ここに抽出したいNo.をリストで入れる

# PaddleOCRの準備
print("PaddleOCRを初期化中...")
ocr = PaddleOCR(use_angle_cls=True, lang='japan')

# PDF読み込み
print(f"PDFを読み込み中: {pdf_path}")
image = pdf_to_cv2_image(pdf_path, dpi=300)

# 全枠線の検出
all_boxes = get_all_boxes(image)
print(f"全 {len(all_boxes)} 個の枠を検出しました。")

# 指定されたNo.ごとに処理
results = {}

for no in target_numbers:
    if no >= len(all_boxes):
        print(f"警告: No.{no} は存在しません（最大 No.{len(all_boxes)-1}）")
        continue

    # 座標取得
    box = all_boxes[no]
    x, y, w, h = box['x'], box['y'], box['w'], box['h']

    # 切り取り
    cropped = image[y:y+h, x:x+w]

    # OCR実行
    print(f"No.{no} を解析中...")
    ocr_result = ocr.ocr(cropped)

    # テキスト抽出
    text = ""
    if ocr_result and ocr_result[0]:
        text = "\n".join([line[1][0] for line in ocr_result[0]])

    results[no] = {"text": text, "img": cropped, "coords": (x, y, w, h)}

# --- 結果の表示 ---
print("\n" + "="*40)
print("一括OCR結果")
print("="*40)

for no, data in results.items():
    print(f"【No.{no}】 座標: {data['coords']}")
    print(f"内容:\n{data['text']}")
    print("-" * 30)

    # 画像を表示（確認用）
    plt.figure(figsize=(8, 2))
    plt.imshow(cv2.cvtColor(data['img'], cv2.COLOR_BGR2RGB))
    plt.title(f"No.{no} Image")
    plt.axis('off')
    plt.show()
```

```python
import cv2
import numpy as np
import fitz  # PyMuPDF
import matplotlib.pyplot as plt
from paddleocr import PaddleOCR
import logging

# 余計な警告ログを非表示にする設定
logging.getLogger("ppocr").setLevel(logging.ERROR)

# --- PDFを画像化する関数 ---
def pdf_to_cv2_image(pdf_path, page_num=0, dpi=300):
    doc = fitz.open(pdf_path)
    page = doc.load_page(page_num)
    zoom = dpi / 72.0
    mat = fitz.Matrix(zoom, zoom)
    pix = page.get_pixmap(matrix=mat)
    img_array = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.h, pix.w, pix.n)
    if pix.n == 4:
        return cv2.cvtColor(img_array, cv2.COLOR_RGBA2BGR)
    elif pix.n == 3:
        return cv2.cvtColor(img_array, cv2.COLOR_RGB2BGR)
    return cv2.cvtColor(img_array, cv2.COLOR_GRAY2BGR)

# ==========================================
# メイン処理
# ==========================================
# PDFのファイル名
pdf_path = "/content/drive/MyDrive/ソリューション開発1&2/フォーマット分類済み/フォーマットM/MPATFS87JNT000219.pdf"

# ★ 複数指定できるように辞書(dict)に変更しました
# 形式: { 枠の番号(No) : (X, Y, W, H) }
# 必要な分だけカンマ区切りで追加できます。以下はNo.5ともう一つのダミー座標の例です。
target_regions = {
    40: (189, 1326, 308, 333),
    41: (501, 1326, 386, 195),
    42: (891, 1326, 1322, 195),
    43: (501, 1525, 386, 65),
    44: (891, 1525, 1322, 65),
    45: (501, 1594, 386, 65),
    46: (891, 1594, 1322, 65),
    # 0: (100, 200, 500, 150),  # ←このように先ほど出力した他のNoの座標を追加してください
    # 3: (x, y, w, h),
}

print("1. PaddleOCRを初期化しています...")
ocr = PaddleOCR(use_angle_cls=True, lang='japan')

print("2. PDFを画像化しています...")
image = pdf_to_cv2_image(pdf_path, dpi=300)

print(f"3. 指定された {len(target_regions)} 箇所の枠を順次処理します...\n")

# 指定した座標の数だけループ処理
for no, coords in target_regions.items():
    x, y, w, h = coords

    print("=" * 50)
    print(f"■ No.{no} の処理を開始 (座標: X:{x}, Y:{y}, W:{w}, H:{h})")
    print("=" * 50)

    # 画像の切り抜き
    cropped_img = image[y:y+h, x:x+w]

    # OCR実行
    result = ocr.ocr(cropped_img)

    # 結果からテキストだけを抽出する
    recognized_text = ""
    if result and result[0]:
        for line in result[0]:
            text = line[1][0]
            recognized_text += text + "\n"

    print(f"【認識結果】\n{recognized_text.strip()}\n")

    # 画像の表示 (各Noごとに表示します)
    plt.figure(figsize=(10, 4))
    plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
    plt.title(f"Cropped Image (No.{no})")
    plt.axis('off')
    plt.show()

print("\n全てのOCR処理が完了しました！")
```

```python
import cv2
import numpy as np
import fitz  # PyMuPDF
import pandas as pd
from paddleocr import PaddleOCR
import logging
import os

logging.getLogger("ppocr").setLevel(logging.ERROR)

def pdf_to_cv2_image(pdf_path, page_num=0, dpi=300):
    doc = fitz.open(pdf_path)
    page = doc.load_page(page_num)
    zoom = dpi / 72.0
    mat = fitz.Matrix(zoom, zoom)
    pix = page.get_pixmap(matrix=mat)
    img_array = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.h, pix.w, pix.n)
    if pix.n == 4:
        return cv2.cvtColor(img_array, cv2.COLOR_RGBA2BGR)
    return cv2.cvtColor(img_array, cv2.COLOR_RGB2BGR)

# ==========================================
# 自動判定ロジックの設定
# ==========================================
pdf_path = "/content/drive/MyDrive/ソリューション開発1&2/フォーマット分類済み/フォーマットM/MPATFS87JNT000219.pdf"

# ★ 変更点：項目名とデータを「ペア」で指定する
# (項目名が書かれているNo, データが書かれているNo)
target_pairs = [
    (41, 42),  # 例: 41を読んで列名にし、42を読んでデータにする
    (43, 44),
    (45, 46)
]

# ★ 各Noの座標データ（本来はテンプレート登録のJSONから読み込むとスマートです）
coords_config = {
    41: (501, 1326, 386, 195),
    42: (891, 1326, 1322, 195),
    43: (501, 1525, 386, 65),
    44: (891, 1525, 1322, 65),
    45: (501, 1594, 386, 65),
    46: (891, 1594, 1322, 65),
}

print("1. PaddleOCRを初期化中...")
ocr = PaddleOCR(use_angle_cls=True, lang='japan')
image = pdf_to_cv2_image(pdf_path, dpi=300)

row_data = {"ファイル名": os.path.basename(pdf_path)}

print("2. 項目名とデータをセットで読み取り、自動判定中...")
for title_no, value_no in target_pairs:
    # --- 項目名（列名）の読み取り ---
    tx, ty, tw, th = coords_config[title_no]
    title_img = image[ty:ty+th, tx:tx+tw]
    title_res = ocr.ocr(title_img)

    # 項目名テキストの抽出
    title_text = "不明な項目"
    if title_res and title_res[0]:
        title_text = "".join([line[1][0] for line in title_res[0]]).replace("\n", "")

    # --- データ（値）の読み取り ---
    vx, vy, vw, vh = coords_config[value_no]
    value_img = image[vy:vy+vh, vx:vx+vw]
    value_res = ocr.ocr(value_img)

    # データテキストの抽出
    value_text = ""
    if value_res and value_res[0]:
        value_text = "".join([line[1][0] for line in value_res[0]])

    # 読み取った「項目名」をキーにして、「データ」を保存
    row_data[title_text] = value_text
    print(f"  [判定完了] 項目名「{title_text}」 ＝ データ「{value_text}」")

# --- Excel出力 ---
df = pd.DataFrame([row_data])
output_file = "auto_judged_results.xlsx"
df.to_excel(output_file, index=False)
print(f"\n✅ Excelを保存しました: {output_file}")
```

```python
import os
import glob
import cv2
import numpy as np
import fitz
import pandas as pd
from paddleocr import PaddleOCR
import logging

logging.getLogger("ppocr").setLevel(logging.ERROR)

def pdf_to_cv2_image(pdf_path, page_num=0, dpi=300):
    doc = fitz.open(pdf_path)
    page = doc.load_page(page_num)
    zoom = dpi / 72.0
    mat = fitz.Matrix(zoom, zoom)
    pix = page.get_pixmap(matrix=mat)
    img_array = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.h, pix.w, pix.n)
    if pix.n == 4:
        return cv2.cvtColor(img_array, cv2.COLOR_RGBA2BGR)
    return cv2.cvtColor(img_array, cv2.COLOR_RGB2BGR)

def parse_ocr_results(ocr_data):
    """OCRの生データから、テキストと扱いやすい座標リストを作成する"""
    parsed = []
    if not ocr_data or not ocr_data[0]:
        return parsed

    for line in ocr_data[0]:
        box = line[0]  # [[x1,y1], [x2,y2], [x3,y3], [x4,y4]]
        text = line[1][0]

        x_coords = [p[0] for p in box]
        y_coords = [p[1] for p in box]
        x_min, x_max = min(x_coords), max(x_coords)
        y_min, y_max = min(y_coords), max(y_coords)
        y_center = (y_min + y_max) / 2

        parsed.append({
            "text": text.strip(),
            "x_min": x_min,
            "x_max": x_max,
            "y_min": y_min,
            "y_max": y_max,
            "y_center": y_center
        })
    return parsed

def get_value_right_of_anchor(parsed_data, anchor_keyword):
    """アンカー（見出し）のすぐ右にあるテキストを取得する"""
    anchor_item = None

    # 1. 画面上の全文字からアンカーを探す（部分一致）
    for item in parsed_data:
        if anchor_keyword in item["text"]:
            anchor_item = item
            break

    if not anchor_item:
        return "" # 見つからない場合は空欄

    candidates = []
    # 2. アンカーの「同じ行」「右側」にあるテキストを探す
    for item in parsed_data:
        if item == anchor_item:
            continue

        # 行の高さの許容範囲（多少の上下のズレは許容する）
        y_tolerance = (anchor_item["y_max"] - anchor_item["y_min"]) * 0.8

        # 同じ行にあるか（Y座標の中心が近いか）
        if abs(item["y_center"] - anchor_item["y_center"]) <= y_tolerance:
            # アンカーより右にあるか（多少の食い込み -20px は許容）
            if item["x_min"] >= anchor_item["x_max"] - 20:
                candidates.append(item)

    # 3. 候補の中で、アンカーに「一番近い（一番左側）」のものを返す
    if candidates:
        candidates.sort(key=lambda x: x["x_min"])
        return candidates[0]["text"]

    return "" # 右側に何も書いていない場合（引渡場所が空欄など）

# ==========================================
# 設定と実行
# ==========================================
folder_path = "/content/drive/MyDrive/ソリューション開発1&2/フォーマット分類済み/フォーマットM"

# 探したい項目名（アンカーキーワード）
# ※ここに書いてある文字をOCRが探して、その右を読み取ります
target_anchors = [
    "搬送先/引渡担当",
    "搬入日時",
    "引渡場所"
]

print("1. PaddleOCRを初期化中...")
ocr = PaddleOCR(use_angle_cls=True, lang='japan')

pdf_files = glob.glob(os.path.join(folder_path, "*.pdf"))
print(f"2. フォルダ内に {len(pdf_files)} 件のPDFが見つかりました。\n")

all_results = []

for i, pdf_path in enumerate(pdf_files, 1):
    filename = os.path.basename(pdf_path)
    print(f"[{i}/{len(pdf_files)}] 処理中: {filename}")

    row_data = {"ファイル名": filename}
    image = pdf_to_cv2_image(pdf_path, dpi=300)

    # ページ全体を一回だけOCRにかけて全文字を抽出する
    result = ocr.ocr(image)
    parsed_data = parse_ocr_results(result)

    # キーワードごとに値を探す
    for anchor in target_anchors:
        value_text = get_value_right_of_anchor(parsed_data, anchor)
        row_data[anchor] = value_text
        print(f"  └ {anchor} : {value_text if value_text else '(空欄)'}")

    all_results.append(row_data)

# Excel保存
df = pd.DataFrame(all_results)
df.to_excel("anchor_ocr_results.xlsx", index=False)
print(f"\n✅ 完了！アンカー方式でExcelファイルを作成しました。")
```

```python
import os
import glob
import cv2
import numpy as np
import fitz
import pandas as pd
from paddleocr import PaddleOCR
import logging

logging.getLogger("ppocr").setLevel(logging.ERROR)

def pdf_to_cv2_image(pdf_path, page_num=0, dpi=300):
    doc = fitz.open(pdf_path)
    page = doc.load_page(page_num)
    zoom = dpi / 72.0
    mat = fitz.Matrix(zoom, zoom)
    pix = page.get_pixmap(matrix=mat)
    img_array = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.h, pix.w, pix.n)
    if pix.n == 4:
        return cv2.cvtColor(img_array, cv2.COLOR_RGBA2BGR)
    return cv2.cvtColor(img_array, cv2.COLOR_RGB2BGR)

# ★ 新規追加：OCR精度を上げるための画像前処理
def enhance_image_for_ocr(img, scale_factor=1.5):
    # 1. 画像の拡大（小さな文字を読みやすくする）
    width = int(img.shape[1] * scale_factor)
    height = int(img.shape[0] * scale_factor)
    resized = cv2.resize(img, (width, height), interpolation=cv2.INTER_CUBIC)

    # 2. グレースケール化
    gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)

    # 3. コントラスト強調（適応的ヒストグラム平坦化）
    # 背景のノイズを抑えつつ、文字だけをくっきりさせる
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    enhanced = clahe.apply(gray)

    # 4. PaddleOCRが処理しやすいように3チャンネル（カラー形式）に戻す
    final_img = cv2.cvtColor(enhanced, cv2.COLOR_GRAY2BGR)

    return final_img

def parse_ocr_results(ocr_data, scale_factor=1.5):
    parsed = []
    if not ocr_data or not ocr_data[0]:
        return parsed

    for line in ocr_data[0]:
        box = line[0]
        text = line[1][0]

        # 拡大した分、座標を元のサイズに戻して記録する
        x_coords = [p[0] / scale_factor for p in box]
        y_coords = [p[1] / scale_factor for p in box]
        x_min, x_max = min(x_coords), max(x_coords)
        y_min, y_max = min(y_coords), max(y_coords)
        y_center = (y_min + y_max) / 2

        parsed.append({
            "text": text.strip(),
            "x_min": x_min,
            "x_max": x_max,
            "y_min": y_min,
            "y_max": y_max,
            "y_center": y_center
        })
    return parsed

def get_value_right_of_anchor(parsed_data, anchor_keyword):
    anchor_item = None
    for item in parsed_data:
        if anchor_keyword in item["text"]:
            anchor_item = item
            break

    if not anchor_item:
        return ""

    candidates = []
    for item in parsed_data:
        if item == anchor_item:
            continue

        y_tolerance = (anchor_item["y_max"] - anchor_item["y_min"]) * 0.8
        if abs(item["y_center"] - anchor_item["y_center"]) <= y_tolerance:
            if item["x_min"] >= anchor_item["x_max"] - 20:
                candidates.append(item)

    if candidates:
        candidates.sort(key=lambda x: x["x_min"])
        return candidates[0]["text"]

    return ""

# ==========================================
# 設定と実行
# ==========================================
folder_path = "/content/drive/MyDrive/ソリューション開発1&2/フォーマット分類済み/フォーマットM"

target_anchors = [
    "搬送先/引渡担当",
    "搬入日時",
    "引渡場所"
]

print("1. PaddleOCRを初期化中...")
# ★ 変更：det_limit_side_lenを大きくして、大きな画像でも縮小させずに読み込ませる
ocr = PaddleOCR(use_angle_cls=True, lang='japan', det_limit_side_len=2048)

pdf_files = glob.glob(os.path.join(folder_path, "*.pdf"))
print(f"2. フォルダ内に {len(pdf_files)} 件のPDFが見つかりました。\n")

all_results = []
scale_factor = 1.5 # 拡大率

for i, pdf_path in enumerate(pdf_files, 1):
    filename = os.path.basename(pdf_path)
    print(f"[{i}/{len(pdf_files)}] 処理中: {filename}")

    row_data = {"ファイル名": filename}

    # 画像化
    original_image = pdf_to_cv2_image(pdf_path, dpi=300)

    # ★ OCRに最適化された画像に変換
    enhanced_image = enhance_image_for_ocr(original_image, scale_factor=scale_factor)

    # OCR実行
    result = ocr.ocr(enhanced_image)
    parsed_data = parse_ocr_results(result, scale_factor=scale_factor)

    for anchor in target_anchors:
        value_text = get_value_right_of_anchor(parsed_data, anchor)
        row_data[anchor] = value_text
        print(f"  └ {anchor} : {value_text if value_text else '(空欄)'}")

    all_results.append(row_data)

df = pd.DataFrame(all_results)
df.to_excel("anchor_ocr_results_high_acc.xlsx", index=False)
print(f"\n✅ 完了！高精度モードでExcelファイルを作成しました。")
```

```python
import os
import glob
import cv2
import numpy as np
import fitz
import pandas as pd
from paddleocr import PaddleOCR
import logging

logging.getLogger("ppocr").setLevel(logging.ERROR)

# --- 1. PDF画像化 ---
def pdf_to_cv2_image(pdf_path, page_num=0, dpi=300):
    doc = fitz.open(pdf_path)
    page = doc.load_page(page_num)
    zoom = dpi / 72.0
    mat = fitz.Matrix(zoom, zoom)
    pix = page.get_pixmap(matrix=mat)
    img_array = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.h, pix.w, pix.n)
    if pix.n == 4:
        return cv2.cvtColor(img_array, cv2.COLOR_RGBA2BGR)
    return cv2.cvtColor(img_array, cv2.COLOR_RGB2BGR)

# --- 2. 枠線検出 ---
def extract_boxes(img, min_area=3000):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15, 5)
    kernel = np.ones((3,3), np.uint8)
    dilated = cv2.dilate(thresh, kernel, iterations=1)
    contours, _ = cv2.findContours(dilated, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    boxes = []
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if min_area < area < (img.shape[0] * img.shape[1] * 0.9):
            epsilon = 0.02 * cv2.arcLength(cnt, True)
            approx = cv2.approxPolyDP(cnt, epsilon, True)
            if len(approx) == 4:
                x, y, w, h = cv2.boundingRect(cnt)
                boxes.append({"x": x, "y": y, "w": w, "h": h})

    # 枠を「上から下、左から右」の順番に並び替えてNo.を振る
    # （y座標のわずかなズレを吸収するため、10ピクセル単位で丸めてソート）
    boxes.sort(key=lambda b: (b['y'] // 10, b['x']))
    return boxes

# --- 3. OCR結果を各枠に当てはめる ---
def parse_ocr_results(ocr_data):
    parsed = []
    if not ocr_data or not ocr_data[0]:
        return parsed

    for line in ocr_data[0]:
        box = line[0]
        text = line[1][0]
        x_coords = [p[0] for p in box]
        y_coords = [p[1] for p in box]
        parsed.append({
            "text": text.strip(),
            "cx": sum(x_coords) / 4, # 文字の中心X座標
            "cy": sum(y_coords) / 4  # 文字の中心Y座標
        })
    return parsed

def map_texts_to_boxes(parsed_data, boxes):
    box_contents = {i: "" for i in range(len(boxes))}
    for item in parsed_data:
        for i, box in enumerate(boxes):
            # 文字の中心がどの枠の中にあるかを判定
            if (box["x"] <= item["cx"] <= box["x"] + box["w"]) and \
               (box["y"] <= item["cy"] <= box["y"] + box["h"]):
                box_contents[i] += item["text"]
                break
    return box_contents

# ==========================================
# 実行部分
# ==========================================
folder_path = "/content/drive/MyDrive/ソリューション開発1&2/フォーマット分類済み/フォーマットM"

target_anchors = [
    "搬送先/引渡担当",
    "搬入日時",
    "引渡場所"
]

print("1. PaddleOCRを初期化中...")
ocr = PaddleOCR(use_angle_cls=True, lang='japan')

pdf_files = glob.glob(os.path.join(folder_path, "*.pdf"))
print(f"2. フォルダ内に {len(pdf_files)} 件のPDFが見つかりました。\n")

all_results = []

for idx, pdf_path in enumerate(pdf_files, 1):
    filename = os.path.basename(pdf_path)
    print(f"[{idx}/{len(pdf_files)}] 処理中: {filename}")

    row_data = {"ファイル名": filename}
    image = pdf_to_cv2_image(pdf_path, dpi=300)

    # 枠を検出してNo.を振る
    boxes = extract_boxes(image)

    # 画像全体を1回だけOCR
    result = ocr.ocr(image)
    parsed_data = parse_ocr_results(result)

    # 各No.の枠に文字を入れる
    box_contents = map_texts_to_boxes(parsed_data, boxes)

    # ★「キーワードが入っている枠の、次のNo」を取得する
    for anchor in target_anchors:
        value_text = ""
        for i in range(len(boxes)):
            if anchor in box_contents[i]:
                # アンカーが見つかったら、その次の枠（i + 1）の中身をデータとする
                if i + 1 < len(boxes):
                    value_text = box_contents[i + 1]
                break # 1つ見つけたら次のアンカーを探しに行く

        row_data[anchor] = value_text
        print(f"  └ {anchor} : {value_text if value_text else '(空欄)'}")

    all_results.append(row_data)

# Excel保存
df = pd.DataFrame(all_results)
df.to_excel("next_box_ocr_results.xlsx", index=False)
print(f"\n✅ 完了！「枠の順番（No+1）」の法則を使ってExcelファイルを作成しました。")
```
